```python
from Bio import SeqIO  # Импортируем SeqIO из Biopython

input_fasta = 'GCF_000004695.1_dicty_2.7_genomic.fna'  # Путь к вашему файлу FASTA
output_dir = 'scaffolds/'  # Папка для вывода файлов скаффолдов

# Создаем функцию для чтения и разделения файла FASTA
def split_fasta_by_scaffolds(input_fasta, output_dir):
    # Чтение FASTA файла
    fasta_sequences = SeqIO.parse(open(input_fasta),'fasta')
    for fasta in fasta_sequences:
        name, sequence = fasta.id, str(fasta.seq)
        with open(output_dir + name + ".fasta", "w") as output_file:
            output_file.write(f">{name}\n{sequence}\n")

# Вызов функции
split_fasta_by_scaffolds(input_fasta, output_dir)
```

```python
%%bash
# Указываем директорию с файлами, которые нужно обработать
DIRECTORY="./scaffolds/"

# Переходим в директорию с файлами
cd $DIRECTORY

# Проходим через все файлы в директории
for file in *; do
    # Проверяем, что это файл (не директория)
    if [ -f "$file" ]; then
        # Запускаем zhunt3 с параметрами и текущим файлом
        ../zhunt3 12 8 12 "$file"
        echo "Processed: $file"  # Выводим сообщение о обработанном файле
    fi
done
```

```python
import pandas as pd
import os

# Указываем путь к вашей директории
directory = "scaffolds"

# Глобальный массив для хранения результатов
results = []

# Проход по файлам в директории
for filename in os.listdir(directory):
    if filename.endswith(".fasta.Z-SCORE"):
        # Удаляем расширение .fasta.Z-SCORE для сохранения в результаты
        clean_filename = filename.rstrip('.fasta.Z-SCORE')

        # Полный путь к файлу
        filepath = os.path.join(directory, filename)
        print("Start {}".format(filepath))
        # Чтение файла в DataFrame, пропускаем первую строку
        df = pd.read_table(filepath, delim_whitespace=True, skiprows=1,
                           names=['Start', 'End', 'None1', 'None2', 'None3', 'Z-score', 'Sequence', 'None4'])

        # Фильтрация DataFrame по условию Z-score >= 400
        filtered_df = df[df['Z-score'] >= 400]

        # Добавление результатов в глобальный массив
        for index, row in filtered_df.iterrows():
            results.append((clean_filename, row['Start'], row['End']))
            print((clean_filename, row['Start'], row['End']))
        print("Finish {}".format(filepath))
# Печатаем результаты
print(results)
```

```python
zhunt_data = pd.DataFrame(results, columns=['Chromosome', 'Start', 'End'])
zhunt_data
```

```python
import re
```

```python
input_file = 'GCF_000004695.1_dicty_2.7_genomic.fna'
pqs = []
pqs_minus = []
pattern="(?:G{3,}[ATGC]{1,7}){3,}G{3,}" #шаблон стандартного квадруплекса на + стренде
pattern_minus="(?:C{3,}[ATGC]{1,7}){3,}C{3,}" #на - стренде
for record in SeqIO.parse(input_file, "fasta"):
    name, sequence = record.id, str(record.seq)
    pqs += [[name, m.start(),m.end(),m.group(0), '+'] for m in re.finditer(pattern,sequence,re.IGNORECASE)] #игнорируем регистр
    pqs_minus += [[name, m.start(),m.end(),m.group(0), '-'] for m in re.finditer(pattern_minus,sequence,re.IGNORECASE)]
```

```python
print(len(pqs))
print(len(pqs_minus))
```
